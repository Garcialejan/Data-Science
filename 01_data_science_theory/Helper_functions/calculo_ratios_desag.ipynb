## 3. Prueba de desagregación completa

- El primer paso consiste en obtener el numero de **viajeros por carretera**, para ello multiplicamos los valores de viajeros por el ratio de viajes por carretera correspondiente

```python
df_carretera = pd.merge(df_test, df_ratios_modal,
                        on = ["P_O", "P_D", "tipo_dia"],
                        how = "left")
```

```python
df_carretera["viajes_carretera"] = (df_carretera.viajes_provinciales * df_carretera.ratio_C)
df_carretera = df_carretera.drop(columns = ["viajes_provinciales", "ratio_C", "ratio_R"])
df_carretera
```

- A continuación obtendremos el número de viajes (vehículos) usando los ratios de ocupación

```python
df_carretera = pd.merge(df_carretera, df_ratios_ocupacion,
                        on = ["tipo_dia"],
                        how = "left")
```

```python
df_carretera["viajes_carretera"] = (df_carretera.viajes_carretera / df_carretera.ratio_ocupacion)
df_carretera = df_carretera.drop(columns = ["tipo_dia", "ratio_ocupacion"])
df_carretera
```

- Una vez obtenidos los viajes por carretera obtenemos los **viajes por municipio**

```python
df_municipio = pd.merge(df_ratios_municipal, df_carretera,
                        on = ["P_O", "P_D", "mes", "day_week"],
                        how = "left").dropna()
```

```python
df_municipio["viajes_mcp_carretera"] = (df_municipio.viajes_carretera * df_municipio.ratio_municipal)
df_municipio = df_municipio.drop(columns = ["ratio_municipal", "viajes_carretera"], axis = 1)
df_municipio
```

- A continuación, obtenemos los **viajes por hora** para cada uno de los municipios, obteniendo de esta forma la distribución diaria de viajes por municipios

```python
df_municipio_hr = pd.merge(df_municipio, df_ratios_horas,
                           on = ["P_O", "P_D", "mes", "day_week"],
                           how = "left").dropna()
```

```python
df_municipio_hr["viajes_mcp_carretera_hr"] = (df_municipio_hr.viajes_mcp_carretera * df_municipio_hr.ratios_horarios)
df_municipio_hr
```

```python
df_municipio_hr = df_municipio_hr[["P_O", "P_D", "M_O", "M_D", "hora", "viajes_mcp_carretera_hr"]]
df_municipio_hr
```

```python
import datetime as dt
import pandas as pd
import re

# Importamos todas las fechas de operaciones especiales
import operaciones_especiales
fechas_oe = operaciones_especiales.fechas_oe_todas

# Se definen las rutas en las que se ecuentran las carpetas con los datos
path_data_folder = "C:/Users/alejandro.gc/Desktop/Proyectos_conjuntos/01154_nuevo_mlr/reparto_ratios/data/ratios/"
path_test_folder = "C:/Users/alejandro.gc/Desktop/Proyectos_conjuntos/01154_nuevo_mlr/reparto_ratios/data/test/"

# Se define la ruta y nombre del archivo de test
test_file_name = "df_20220802_provincial_diaria.csv"
path_datos_test = path_test_folder + test_file_name

# Se definen las rutas donde se encuentran los ratios
path_ratios_modo = path_data_folder + "ratios_desagregacion_modal_extra_prov.csv"
path_ratios_ocupacion = path_data_folder + "ratios_ocupacion.csv"
path_ratios_municipales = path_data_folder + "ratios_desagregacion_municipal.csv"
path_ratios_horarios = path_data_folder + "ratios_desagregacion_horaria_final.csv"


def carga_datos_test(path_datos_test: str,
                     file_name: str,
                     sep: str = "|", 
                     columns_drop: tuple[str] = ["o_pro_name", "d_pro_name"],
                     new_names: tuple[str] = ["P_O", "P_D", "viajes_provinciales"],
                     formato_fecha: str = "%Y-%m-%d",
                     entre_semana: list[int] = [0, 1, 2, 3, 4]) -> tuple[pd.DataFrame, str]:
    '''
    Función para cargar los datos de prueba, eliminar las columnas
    que a priori no son necesarias, eliminar los viajes intraprov
    y renombrar las columnas definitivas. Por último extrae la fecha
    en función del nombre del archivo. Una vez obtenida la fecha, se
    obtienen las columnas de mes y day_week necesarias

    Args:
    sep: definimos el separados del csv
    path_datos_test: se corresponde con la ruta en la que se encuentran los datos
    file_name: nombre del archivo en el que se encuentran los datos
    '''
    # Cargamos el df y renombramos las columnas
    df_test = pd.read_csv(path_datos_test, sep = sep, index_col= 0, low_memory=False)\
        .drop(columns = columns_drop, axis = 1)
    
    rename_dict = dict(zip(list(df_test.columns), new_names))
    df_test = df_test.rename(columns = rename_dict)

    # Eliminamos los viajes intraprovinciales
    df_test = df_test.loc[df_test.P_O != df_test.P_D]

    # Incroporamos la fecha y extraemos mes, day_week y tipo_dia
    match = re.search(r'(\d{4})(\d{2})(\d{2})', file_name)
    if match:
        year, month, day = match.groups()
        fecha_formateada = f"{year}-{month}-{day}"

    # Definimos las nuevas columnas 
    df_test["fecha"] = pd.to_datetime(fecha_formateada, format = formato_fecha)
    df_test["mes"] = df_test["fecha"].dt.month
    df_test["day_week"] = df_test["fecha"].dt.weekday

    if fecha_formateada in fechas_oe:
        df_test["tipo_dia"] = "operacion_especial"
    else:
        df_test["tipo_dia"] = df_test.day_week.apply(lambda x: "entre_semana" if x in entre_semana else "fin_semana")

    df_test = df_test.drop(columns = "fecha", axis = 1)
    
    return df_test, fecha_formateada


def carga_ratios_general(path: str) -> pd.DataFrame:
    '''
    Función que permite obtener los ratios de desagregación
    a partir de un path especcificado.

    Args:
    path: ruta del archivo con los valores de los ratios
    '''
    df = pd.read_csv(path, sep = ",", index_col= 0, low_memory=False)
    return df


def carga_ratios_municipal(path: str,
                           fecha:str,
                           formato_fecha: str = "%Y-%m-%d") -> pd.DataFrame:
    '''
    Función que permite obtener los ratios de desagregación
    municipal. Se ha creado una función especial para realizar
    un filtrado por fecha.

    Args:
    path: ruta del archivo con los valores de los ratios
    fecha: fecha de la que se toman los datos
    '''
    fecha_datetime = pd.to_datetime(fecha, format=formato_fecha)
    mes_test = fecha_datetime.month
    day_week_test = fecha_datetime.weekday()
    df = pd.read_csv(path, sep = ",", index_col= 0, low_memory=False)

    # Nos quedamos sólo con los ratios correspondientes al
    # mes y day_week que queremos estudiar
    df = df[df.mes.eq(mes_test)&df.day_week.eq(day_week_test)]

    return df

def merge_dataframes_left(df_left: pd.DataFrame,
                          df_right: pd.DataFrame,
                          columns_merge: list[str]) -> pd.DataFrame:
    '''
    Función simple que fusiona dos dataframes, el
    principal es el de la izquierda.

    Args
    df_left: dataframe principal sobre el que se fusoiona
    df_right. dataframe cuyos datos son incorporados
    columns_merge: columnas a prtir de las cuales se realiza
    la fusión
    '''
    df_merge = pd.merge(df_left,
                        df_right,
                        on = columns_merge,
                        how = "left")
    return df_merge
    

def calculo_desagregacion(df_test: pd.DataFrame,
                          df_ratios_desagregacion: pd.DataFrame,
                          new_column_name:str,
                          column_viajes: str,
                          column_ratios:str,
                          columns_to_merge: list[str],
                          columns_to_drop: list[str]) -> pd.DataFrame:
    
    df_result = merge_dataframes_left(df_test,
                                      df_ratios_desagregacion,
                                      columns_to_merge)
    
    df_result[new_column_name] =  df_result[column_viajes] * df_result[column_ratios]
    df_result = df_result.drop(columns = columns_to_drop)

    return df_result


def main():
    '''
    Función que realiza la desagregación de viajeros. Partiendo
    de datos de viajeros provinciales se obtiene: número de viajes
    (vehículos) por carretera, municipio y hora.
    '''
    # Cargamos los datos que queremos desagregar
    df_test, fecha_test = carga_datos_test(path_datos_test = path_datos_test,
                                           file_name= test_file_name)

    # Cargamos los ratios calculados
    df_ratios_modo = carga_ratios_general(path_ratios_modo)
    df_ratios_ocupacion = carga_ratios_general(path_ratios_ocupacion)
    df_ratios_municipal = carga_ratios_municipal(path_ratios_municipales, fecha_test)
    df_ratios_hora = carga_ratios_general(path_ratios_horarios)

    # Obtenemos, de forma secuencial, los viajes desagregados
    # Paso 1: obtenemos los viajes por carretera
    df_carretera = calculo_desagregacion(df_test,
                                         df_ratios_modo,
                                         new_column_name = "viajes_carretera",
                                         column_viajes = "viajes_provinciales",
                                         column_ratios = "ratio_C",
                                         columns_to_merge = ["P_O", "P_D", "tipo_dia"],
                                         columns_to_drop = ["viajes_provinciales", "ratio_C", "ratio_R"])
    
    # Paso 2: obtenemos el número de vehiculos
    df_vehiculo_carretera = calculo_desagregacion(df_carretera,
                                                  df_ratios_ocupacion,
                                                  new_column_name = "vehiculos_carretera",
                                                  column_viajes = "viajes_carretera",
                                                  column_ratios = "ratio_ocupacion",
                                                  columns_to_merge = ["tipo_dia"],
                                                  columns_to_drop = ["tipo_dia", "viajes_carretera"])
    
    # Paso 3: desagregación por municpios
    df_municipio = calculo_desagregacion(df_ratios_municipal,
                                         df_vehiculo_carretera,
                                         new_column_name = "vehiculos_mcp_carretera",
                                         column_viajes = "vehiculos_carretera",
                                         column_ratios = "ratio_municipal",
                                         columns_to_merge = ["P_O", "P_D", "mes", "day_week"],
                                         columns_to_drop = ["ratio_municipal", "vehiculos_carretera"])
    
    # Paso 4: Desagregación por horas
    df_municipio_hr = calculo_desagregacion(df_municipio,
                                            df_ratios_hora,
                                            new_column_name = "vehiculos_mcp_carretera_hr",
                                            column_viajes = "vehiculos_mcp_carretera",
                                            column_ratios = "ratios_horarios",
                                            columns_to_merge = ["P_O", "P_D", "mes", "day_week"],
                                            columns_to_drop = ["ratios_horarios", "vehiculos_mcp_carretera"])
    
    # Solo tomamos las columnas que nos interesan
    final_columns = ["P_O", "P_D", "M_O", "M_D", "hora", "vehiculos_mcp_carretera_hr"]
    df_municipio_hr = df_municipio_hr[final_columns]

    return df_municipio_hr
```

```python
df_municipio_hr = main()
```
